```python
import numpy as np
import scipy as sp
import math
import matplotlib.pyplot as plt
```

```python
dopAndStark = np.loadtxt('DopplerAndStark.csv', delimiter=',')
fineDopAndStark = np.loadtxt('FineDopplerAndStark.csv', delimiter=',')
HRSJump = np.loadtxt('HRSJump.csv', delimiter=',')
MHRS2 = np.loadtxt('MHRS2.csv', delimiter=',')
Ramseypi2 = np.loadtxt('Ramseypi2.csv', delimiter=',')
Ramsey3pi2 = np.loadtxt('Ramsey3pi2.csv', delimiter=',')
Ramsey5pi2 = np.loadtxt('Ramsey5pi2.csv', delimiter=',')
Ramsey7pi2 = np.loadtxt('Ramsey7pi2.csv', delimiter=',')
newRamsey5pi2 = np.loadtxt('NewRamsey5pi2.csv', delimiter=',')
```

```python
HRSJumpFront = np.loadtxt('HRSJumpFront.csv', delimiter=',')
Ramsey3pi2New = np.loadtxt('Ramsey3pi2Matched.csv', delimiter=',')
fineRamsey3pi2 = np.loadtxt('FineRamsey3pi2Matched.csv', delimiter=',')
```

```python
pts = 20
nbarfs = np.linspace(0, 40, pts)
fineNbarfs = np.linspace(0,15, 50)
```

```python
plt.figure(figsize=(14,10), dpi=500)
plt.plot(nbarfs, dopAndStark*1e3, color="orange", linewidth=6,label="Doppler + DC Stark")
plt.plot(nbarfs, Ramseypi2*1e3, linewidth=2, color="tab:blue", label=r"RS $\pi/2+\pi/2$")

plt.plot(nbarfs, Ramsey3pi2New*1e3, linewidth=2, color="red", label=r"RS $\pi/2 + 3\pi/2$")
plt.plot(nbarfs, newRamsey5pi2*1e3, linewidth=2, color="cyan", label=r"RS $\pi/2+5\pi/2$")
plt.plot(nbarfs, Ramsey7pi2*1e3, linewidth=2, color="firebrick", label=r"RS $\pi/2+7\pi/2$")

plt.plot(nbarfs, HRSJump*1e3,color="magenta", linewidth=2, label=r"HRS180 $\pi/2+3\pi/2$")
plt.plot(nbarfs, HRSJumpFront*1e3, color='darkviolet', linewidth=2, label=r"HRS180 $3\pi/2+\pi/2$")


#plt.plot(nbarfs, MHRS2*1e3)
#plt.plot(nbarfs, Ramseypi2*1e3)
#plt.plot(nbarfs, -Ramsey3pi2*1e3/15)
#plt.plot(nbarfs, Ramsey5pi2*1e3)
#plt.ylim([0,3])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel(r"Final $\bar{n}$",fontsize="25")
plt.ylabel("Clock Shift (mHz)",fontsize="25")
plt.legend(fontsize="21")
plt.show()
```
